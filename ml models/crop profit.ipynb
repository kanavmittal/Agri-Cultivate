 ## Model Building

```python
X=df[['State','Crop','Support price','CC2','CP2']]
Y=df['profit_label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,shuffle=True,random_state=22)
```

```python
X_train.head()
```

```python
y_train
```

```python
len(X_test)
```

```python
y_test
```

```python
#os1=RandomOverSampler(sampling_strategy=0.50)
#X_test_os,y_test_os=os1.fit_resample(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_train)
```

```python
weights=np.linspace(0,0.8,100)
```

```python
grid={'penalty':['l2'],
      'class_weight':[{0:1-x,1.0:x} for x in weights],
       "max_iter":[5000]
     }
```

```python
lr_clf=GridSearchCV(estimator=lr,param_grid=grid,cv=10)
```

```python
lr_clf.get_params().keys()
```

```python
lr_clf.fit(X_train,y_train)
```

```python
lr_clf.best_params_
```

```python
y_pred=lr_clf.predict(X_train)
```

```python
y_pred_test=lr_clf.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred_test))
print("The test accuracy is :{}".format(accuracy_score(y_test,y_pred_test)))
print("The train accuracy is :{}".format(accuracy_score(y_train,y_pred)))
```

```python
print(classification_report(y_test,y_pred_test))
```
